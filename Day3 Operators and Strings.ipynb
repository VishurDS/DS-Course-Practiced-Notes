# **Equality Operators**

1. is (_pointing to same the object_)

meaning is the they will have same memory address
2. is not ((_not pointing to same the object_)
3. == (equal to) (have same value)
4. != (is not equal to)

```python
lst1 = [1,2,3,4]
lst2 = [1,2,3,4]
```

**Checking the memory address of the object**

**id()**---> function to check the memory address

```python
print(id(lst1))
print(id(lst2))
```

So above two objects has the different memory address, hence they are not pointing to same obeject

```python
lst1 is lst2 #hence the result is False here as both lst1 and lst 2 are not pointing to same address although values are same
```

```python
lst1 is not lst2
```

```python
lst1 == lst2   #as both contains the same values/elements hence we are getting True here
```

```python
#now
lst1 = lst2
```

```python
print(id(lst1))   #now both lst1 and lst2 are pointing to the same address
print(id(lst2))
```

```python
lst1 is lst2   #now we are getting true as both are poitning to the same object same memory location
```

length and count are not same

That's because empty strings are considered to exist between all the characters of a string; for a string length 2, there are 3 empty strings; one at the start, one between the two characters, and one at the end.

Generally speaking, the empty string will match at all positions in a given string, including right at the start and end, so the count should always be the length plus 1

```python
size = int(input("Enter the triangle length"))
for i in range(size):
  for j in range(size-i):
    print("", end=" ")
  for k in range(i+1):
    print("*",end = "")
  for m in range(k-1):
    print("*", end="")
  print()
```

```python
## Solution
n = int(input("Enter the triangle length"))
for i in range(n):
  print(" "*(n-i), end="")
  for j in range(i*2+1):
    print("*", end = "")
  print()
```
