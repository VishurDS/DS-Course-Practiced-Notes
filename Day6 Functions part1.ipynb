If we want to pass any number of arguments in the form of key value pair then we use **kwargs. Means way to deal with dictionary  (key word arugment)

Make a note that Double asterisk is important.

Function will return dictionary here.

Incase of single asterisk function returned the tuple.

```python
def test5(**kwargs):
  return kwargs
```

```python
test5() #it will return the empty dictionary
```

```python
type(test5()) #here type of test5 is dictionary datatype
```

```python
test5(a = 35, b = 23, c = [1,2,3,4], d = ("Vish", "Data Analyst"))  # we got result in the form of key value pair
```

```python
def test6(**key_value_pairs):
  return key_value_pairs
```

```python
test6(a = 35, b = 23, c = [1,2,3,4], d = ("Vish", "Data Analyst"))
```

Here two asterisk (**) are important because kwargs is also not a keyword in Python.
You can use anything in place of kwargs.

Function will return the result in key value pair i.e. dictionary

```python
#extract all the key value pair where value is of list type

def test7(**blah_blah):
  for i in blah_blah.keys():
    if type(blah_blah[i]) == list:
      return i, blah_blah[i]
```

```python
test7(a = 35, b = 23, c = [1,2,3,4], d = ("Vish", "Data Analyst"))
```

Passing both single asterisk and double asterisk

```python
def test8(*args, **kwargs):  #means it will take n number of arguments and n number of key value pairs
  return args, kwargs
```

 A Generator in Python is a function that returns an *iterator* using the **Yield** keyword

 Gives record one by one

 Example: range()

```python
range(1,10) #generating the data
```

Means we want to generate the range of data

Does not store data
Generates data one by one data give it

Saves memory as well

For efficient memory utilization.

Example is Fibonacci Series


```python
#Creating a generator function

def test_fib(n):
  a,b = 0, 1
  for i in range(n):
    yield a #this is important
    a, b = b, a+b #logic
```

```python
test_fib(10)  #so this is generator function
```

```python
range(10)
```

How the generator function works


```python
for i in test_fib(10):
  print(i)
```

```python
#Prime numbers program
#Source Programiz
# Python program to display all the prime numbers within an interval

lower = 10
upper = 20

print("Prime numbers between", lower, "and", upper, "are:")

for num in range(lower, upper + 1):
   # all prime numbers are greater than 1
   if num > 1:
       for i in range(2, num):
           if (num % i) == 0:
               break
       else:
           print(num)
```

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
          return False
    return True


def generate_prime(lower, upper):
  for x in range(lower, upper+1):
    if is_prime(x):
      yield (x)
```

```python
prime_num = iter(generate_prime(2, 1000))
for i in range(1, 21):
  print(next(prime_num))
```
